```python
import librosa
import pandas as pd
import numpy as np
import os
import math
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
folder_serial = 'f';
folder_path = fr'../02_Data Cleaning/Dataset_ready/training-{folder_serial}';
list_of_files = os.listdir(folder_path);
list_of_files = [sound_files for sound_files in list_of_files if '.wav' in sound_files];
reference_file = fr'../01_Dataset Exploration/Dataset/training-{folder_serial}/REFERENCE.csv';
reference = pd.read_csv(reference_file);
```

```python
def chromatic_features(y , sr):
    D = librosa.stft(y)
    D_harmonic, D_percussive = librosa.decompose.hpss(D)
    D_harmonic4, D_percussive4 = librosa.decompose.hpss(D, margin=4)
    y_harmonic4 = librosa.istft(D_harmonic4, length=len(y))
    y_percussive4 = librosa.istft(D_percussive4, length=len(y))
    chroma_stft_12_h = librosa.feature.chroma_stft(y=y_harmonic4, sr=sr, n_chroma=12)
    #print("the length after chroma_stft is",chroma_stft_12.shape);
    chroma_stft_12_p = librosa.feature.chroma_stft(y=y_percussive4, sr=sr, n_chroma=12)
    return (chroma_stft_12_h , chroma_stft_12_p)
    #print("the length after chroma_stft is",chroma_stft_12.shape);
```

```python
df = pd.DataFrame()
```

```python
for x in range(len(list_of_files)):
    soundtrack = list_of_files[x];
    path = rf'../02_Data Cleaning/Dataset_ready/training-{folder_serial}/{soundtrack}';
    y , sr = librosa.load(path);
    time = librosa.get_duration(y = y , sr = sr);
    time = math.floor(time);
    for z in range(time-1):
        y1 , sr1 = librosa.load(path , offset = z+1, duration = 1)
        a = b = {};
        h , p = chromatic_features(y1, sr1);
        for i in range(len(h)):
            a[f'harmonic{i+1}'] = np.mean(h[i]);
            b[f'percussiv{i+1}'] = np.mean(p[i]);
        res = a | b;
        data = pd.DataFrame(a , index = [0])
        data['class'] = reference.iloc[x , 1];
        lst = [df , data]
        df = pd.concat(lst , axis = 0)
#         extracted_features_list.append([mfcc_scaled_features , reference.iloc[x , 1]])
    print(soundtrack, end = " ")
```
