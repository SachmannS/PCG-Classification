```python
import pandas as pd;
import librosa as lb;
import os
import math
```

```python
alpha = 'f';
data = pd.read_csv(f'./Extracted_features/Extracted_features_modifies_mfcc44/trainnig-{alpha}.csv')
```

```python
folder_path = fr'../02_Data Cleaning/Dataset_ready/training-{alpha}';
list_of_files = os.listdir(folder_path);
list_of_files = [sound_files for sound_files in list_of_files if '.wav' in sound_files];
```

```python
l = []
for x in range(len(list_of_files)):
    soundtrack = list_of_files[x];
    path = rf'../02_Data Cleaning/Dataset_ready/training-{alpha}/{soundtrack}';
    y , sr = lb.load(path);
    time = lb.get_duration(y = y , sr = sr);
    time = math.floor(time);
    for z in range(time-1):
        l.append(soundtrack)
    print(soundtrack, end = " ")
```

```python
data['Sample'] = l
```
